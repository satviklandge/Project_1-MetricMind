```python
# Cell 4 — Connect to PostgreSQL  ✅ CORRECTED
import urllib.parse
from sqlalchemy import create_engine, text
DB_USER     = "postgres"
DB_PASSWORD = "Satvik@19"
DB_HOST     = "localhost"
DB_PORT     = "5432"
DB_NAME     = "metricmind"
# quote_plus turns "Satvik@19" → "Satvik%4019" so the @ doesn't break the URL
encoded_pw  = urllib.parse.quote_plus(DB_PASSWORD)
# Note: f"..." makes it an f-string so variables get substituted
conn_string = f"postgresql+psycopg2://{DB_USER}:{encoded_pw}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine = create_engine(conn_string)
with engine.connect() as conn:
    conn.execute(text("SELECT 1"))
    print("✅ Connected to PostgreSQL successfully!")
```
