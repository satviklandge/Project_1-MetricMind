## Check Duplicate Records


```python
print(df.duplicated().sum())
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

## Validate Business Rules

### Revenue

```python
check = df["Quantity"] * df["Unit_Price"]

print((check == df["Revenue"]).all())
```

### Profit 

```python
check = df["Revenue"] - df["Total_Cost"]

print((check.round(2) == df["Profit"]).all())
```

### Margin 

```python
margin = ((df["Profit"] / df["Revenue"]) * 100).round(2)

print((margin == df["Margin"]).all())
```

## European shipping costs increased.

-> Because of that:

->Revenue remained stable
->Shipping cost increased
->Total cost increased
->Profit decreased
->Margin dropped

```python
df["Quarter"] = df["Order_Date"].dt.quarter

europe = df[df["Region"] == "Europe"]

print(
    europe.groupby("Quarter")[["Shipping_Cost", "Margin"]].mean()
)
```
