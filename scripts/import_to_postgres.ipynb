```python
from sqlalchemy import create_engine
import pandas as pd

df = pd.read_csv("data/processed/sales_data_clean.csv")

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

engine = create_engine(
    "postgresql+psycopg2://postgres:Satvik%4019@localhost:5432/metricmind"
)

df.to_sql(
    "sales_data",
    engine,
    if_exists="append",
    index=False
)

print("Data imported successfully!")
```
