```python
import pandas as pd 
```

```python
## Load dataset 
df = pd.read_excel("data/raw/sales_data.csv")
df.head()
df.shape
```

## Dataset information 

```python
df.info()
```

## Statistical Summary

```python
df.describe()
```

## Check Missing Value

```python
df.isnull().sum()
```

 ## Check Duplicate Records

```python
df.duplicated().sum()
```

```python
## convert date
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

## Validate Bussiness Rules

```python
### Revenue 

check = df["Quantity"] * df["Unit_Price"]
print((check == df ["Revenue"]).all())
```

```python
### Profit

check = df["Revenue"] - df["Total_Cost"]
print((check.round(2) == df["Profit"]).all())
```

```python
### Margin 

margin = ((df["Profit"] / df["Revenue"]) * 100).round(2)
print((margin == df["Margin"]).all())
```

## European shipping costs increased

# Because of that:
->Revenue remained stable
->Shipping cost increased
->Total cost increased
->Profit decreased
->Margin dropped

```python
df["Quarter"] = df["Order_Date"].dt.quarter

europe = df[df["Region"] == "Europe"]

print(
    europe.groupby("Quarter")[["Shipping_Cost", "Margin"]].mean()
)
```

## Save Clean Dataset 


```python
df.to_csv(
    "data/processed/sales_data_clean.csv",
    index=False
)
```
